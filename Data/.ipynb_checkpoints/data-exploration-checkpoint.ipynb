# Importing liberaries

#### Will run this code in Jupyter because in order to show context.show it requires html which is can be plot here
Each compenent has an Input and an output
The Input usually the required compenent's out put

```python
import tfx
from tfx.components import CsvExampleGen
from tfx.components import StatisticsGen
import os
from tfx.proto import example_gen_pb2
from tfx.v1 import proto
from tfx.orchestration.experimental.interactive.interactive_context import InteractiveContext
```

## Defining Our working directory

```python
_pip_line_root = '../pip_line' # using ".." to go back one folder #TODO make it in config file
if not os.path.exists(_pip_line_root):
    os.makedirs(_pip_line_root)
    
prep_data_dir = "Prep-Data"#TODO make it in config file
```

# Creating InteractiveContext 

```python
context = InteractiveContext(
    pipeline_name='rec_sys_imdb',
    pipeline_root=_pip_line_root # That's the folder where all components will be availbe to use for the pipleine
)
```

# Creating Example Gen component

## The first step to ingest our data
This code splits our data for train, eval and test at a ratio of 8:1:1

```python
output = proto.Output(
    split_config=example_gen_pb2.SplitConfig(splits=[
        proto.SplitConfig.Split(name="train", hash_buckets=8),
        proto.SplitConfig.Split(name="eval", hash_buckets=1),
        proto.SplitConfig.Split(name="test", hash_buckets=1)
    ])
)

example_gen = CsvExampleGen(input_base=prep_data_dir, output_config=output)
```

```python
context.run(example_gen)
```

```python
artifact = example_gen.outputs['examples'].get()[0]
print(artifact.split_names, artifact.uri)
```

# Creating statistics gen
 In order to create statistics gen we have to use our example_gen examples

```python
statistics_gen = StatisticsGen(
    examples=example_gen.outputs['examples'])
context.run(statistics_gen)
```

```python
context.show(statistics_gen.outputs['statistics'])
```

# Creating Schemagen

```python
schema_gen = tfx.components.SchemaGen(
    statistics=statistics_gen.outputs['statistics'],
    infer_feature_shape=False)
context.run(schema_gen)
```

```python
context.show(schema_gen.outputs['schema'])
```

# Creating Validator
We are looking if there are anomalies or not, and if there are anomalies? Do we want to treat them as Anomalies or tell our compenent it's not an anomaly.

```python
example_validator = tfx.components.ExampleValidator(
    statistics=statistics_gen.outputs['statistics'],
    schema=schema_gen.outputs['schema'])
context.run(example_validator)
```

```python
context.show(example_validator.outputs['anomalies'])
```
